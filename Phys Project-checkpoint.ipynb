# The Electric field at the antenna
To find the electric field within the HERA antenna the total distance that the signal has to travel from the source to the feed needs to be found first, a point *p* is chosen on the dish at co-ordinates $(x,y)$ and a point perpindicular to this point at the source with co-ordinates $(0,H_{f})$. The feed co-ordinates are given as $(0,h_{f})$, where $h_{f}$ is the height of the feed. $H_{f}$ is found using the far-field approximation $\frac{2D^{2}}{\lambda}$.

The distance from source to feed is given as $$ L=\sqrt{x^{2}+(y-h_{f})^{2}} + H_{f} - y $$ 
The distance *L* is used to find the phase $\phi$ $$ \phi = \frac{2\pi{L}}{\lambda} $$ The Electric Field at the feed is $$ dE=E_{0}e^{i\phi}ds $$

Where $ds$ is the path area on the antenna and is given by 
$$ds=dx\sqrt{1+\frac{x^{2}}{4h_{f}^{2}}}$$

The total Electric field at the feed of the antenna is given by $$ E_{Total}=\int_\frac{-D}{2}^\frac{D}{2} E_{0}e^{i\phi}ds$$ where *D* is the diameter of the antenna dish


 *Figure 1: A simple diagram of the HERA antenna showing the co-ordinates of the source at $(0,H_{f})$, a point p on the antenna dish at $(x,y)$ and the feed at  $(0,h_{f})$.*

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.integrate as integrate
```

```python
#Functions of the equations
D = 14.    # Diameter of dish in meters
wavelen = 2. # frequency of 150Mhz in metres
e = math.e   # eulers number
dx = 0.
dy=0.
E0 = 1.

#Make y a function
h_feed = 4.48 #feed height in meters found from eq: focal length/diameter radio = 0.32 (HERA, DeBoer)

dfeed = lambda h_feed, dx,dy: np.sqrt((dx**2)+(h_feed-dy)**2) #Feed height with positional variation

H_far =lambda D, wavelen: (2*D**2)/wavelen
y = lambda x, h_feed,dx,dy: (x**2)/(4*dfeed(h_feed, dx,dy))
```

```python
x = np.linspace(-7,7,15)
plt.plot(x,y(x,h_feed,dx,dy))
```

```python
L =  lambda x,h_feed,D,wavelen,dx,dy: np.sqrt(x**2 + (y(x,h_feed,dx,dy)-dfeed(h_feed, dx,dy))**2) + H_far(D,wavelen)-y(x,h_feed,dx,dy)

plt.plot(x,L(x,h_feed,D,wavelen,dx,dy))
```

```python
phi =lambda wavelen, h_feed,x,D,dx,dy: (2*(np.pi)*(L(x,h_feed,D,wavelen,dx,dy)/wavelen))
plt.plot(x,phi(wavelen,h_feed,x,D,dx,dy))
```

```python
ds= lambda x,h_feed,dx,dy: np.sqrt(1+(x**2/((dfeed(h_feed, dx,dy))**2)*4))
plt.plot(x,ds(x,h_feed,dx,dy))
print(ds(x,h_feed,dx,dy))
```

```python
#dA = lambda x,h_feed,dx,dy: abs(2*(np.pi)*x*(d_s(x,h_feed,dx,dy)))
#plt.plot(x,dA(x,h_feed,dx,dy))
```

```python
dEreal = lambda x, h_feed,wavelen,D,dx,dy: ds(x, h_feed,dx,dy)*(np.cos(phi(wavelen,h_feed,x,D,dx,dy)))
plt.plot(x,dEreal(x, h_feed,wavelen,D,dx,dy))
print(dEreal(x, h_feed,wavelen,D,dx,dy))
```

```python
dEimg = lambda x, h_feed,wavelen,D,dx,dy: ds(x, h_feed,dx,dy)*(np.sin(phi(wavelen,h_feed,x,D,dx,dy)))
plt.plot(x,dEimg(x, h_feed,wavelen,D,dx,dy))
print(dEimg(x, h_feed,wavelen,D,dx,dy))
```

```python
Etotal_real = integrate.quad(dEreal,-D/2,D/2, args=(dfeed(h_feed,dx,dy),wavelen,D,dx,dy))
print(Etotal_real)
```

```python
Etotal_img = integrate.quad(dEimg,-D/2,D/2, args=(dfeed(h_feed,dx,dy),wavelen,D,dx,dy))
print(Etotal_img)
```

```python
Etotal = Etotal_real[0]+1j*Etotal_img[0]
err = Etotal_real[1]+1j*Etotal_img[1]
print(Etotal)
print(err)
```

# With Theta
In this section we show how the eletric field changes with angle $\theta$ here all equations used are the same,but know we want to show for different angles of $\theta$ where $\theta$ ranges $180^{o}$ from $-90^{o}$ to $90^{o}$ .

The total distance from source to feed is now $$ L=\sqrt{x^{2}+(y-h_{f})^{2}} + |x\sin\theta+y\cos\theta-H_{f}| $$


```python
#Functions of the equations
D = 14.    # Diameter of dish in meters
wavelen = 0.21 # 21 cm in meters Frequncy range 100MHz to 200MHz, make an array for lambda
e = math.e   # eulers number
dx = 0.
dy=0.
E0 = 1.
theta = (np.linspace(-np.deg2rad(90),np.deg2rad(90),200))

#Make y a function
h_feed = 4.48 #feed height in meters found from eq: focal length/diameter radio = 0.32 (HERA, DeBoer)

dfeed = lambda h_feed, dx,dy: np.sqrt((dx**2)+(h_feed-dy)**2) #Feed height with positional variation

H_far =lambda D, wavelen: (2*D**2)/wavelen
y = lambda x, h_feed,dx,dy: (x**2)/(4*dfeed(h_feed, dx,dy))
```

```python
x = np.linspace(-7,7,200)
plt.plot(x,y(x,h_feed,dx,dy))
```

```python
L =  lambda x,h_feed,D,wavelen,dx,dy,theta: np.sqrt(x**2 + (y(x,h_feed,dx,dy)-dfeed(h_feed, dx,dy))**2) + abs(x*np.sin(theta)+y(x,h_feed,dx,dy)*np.cos(theta)-H_far(D,wavelen))

plt.plot(x,L(x,h_feed,D,wavelen,dx,dy,theta))
```

```python
phi =lambda wavelen, h_feed,x,D,dx,dy,theta: (2*(np.pi)*(L(x,h_feed,D,wavelen,dx,dy,theta)/wavelen))
plt.plot(x,phi(wavelen,h_feed,x,D,dx,dy,theta))
```

```python
ds= lambda x,h_feed,dx,dy: np.sqrt(1+(x**2/((dfeed(h_feed, dx,dy))**2)*4))
plt.plot(x,ds(x,h_feed,dx,dy))
#print(d_s(x,h_feed,dx,dy))
```

```python
dEreal = lambda x, h_feed,wavelen,D,dx,dy,theta: ds(x, h_feed,dx,dy)*(np.cos(phi(wavelen,h_feed,x,D,dx,dy,theta)))
#plt.plot(x,dEreal(x, h_feed,wavelen,D,dx,dy,nx,theta))
#print(dEreal(x, h_feed,wavelen,D,dx,dy,theta))
# dET = dEreal+1jdEimg
```

```python
plt.subplots(figsize=(20, 5))


plt.subplot(1,2,1)
plt.plot(theta,dEreal(x, h_feed,wavelen,D,dx,dy,theta))
plt.title('$\\theta$(rad) vs $dE_{Real}$')
plt.xlabel('$\\theta$(rad)')
plt.ylabel('$dE_{Real}$')

plt.subplot(1,2,2)
plt.plot(theta,dEimg(x, h_feed,wavelen,D,dx,dy,theta))
plt.title('$\\theta$(rad) vs $dE_{Imaginary}$')
plt.xlabel('$\\theta$(rad)')
plt.ylabel('$dE_{Imaginary}$')
#plt.savefig('theta_vs_dE.jpg')
```

```python

Etotal_real =lambda  h_feed,wavelen,D,dx,dy,theta: integrate.quad(dEreal,-D/2,D/2, args=(dfeed(h_feed,dx,dy),wavelen,D,dx,dy,theta),limit=100)

Etotal_img = lambda h_feed,wavelen,D,dx,dy,theta: integrate.quad(dEimg,-D/2,D/2, args=(dfeed(h_feed,dx,dy),wavelen,D,dx,dy,theta),limit=100)
```

```python
Etotal = np.zeros(np.shape(theta))
for i in range(theta.size):

    #print(Etotal_real( h_feed,wavelen,D,dx,dy,theta[i])[0],Etotal_img(h_feed,wavelen,D,dx,dy,theta[i])[0])
    #Etotal[i]= Etotal_real+1j*Etotal_img
    #print(Etotal_real,Etotal_img,i,theta[i])
    Etotal[i] = abs(Etotal_real( h_feed,wavelen,D,dx,dy,theta[i])[0]+1j*Etotal_img(h_feed,wavelen,D,dx,dy,theta[i])[0])
```
